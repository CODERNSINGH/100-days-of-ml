### Opening a from URL
* we don't have to learn this it will be avilable online
* let's suppose link https://people.sc.fsu.edu/~jburkardt/data/csv/addresses.csv

```python
import requests
from io import StringIO
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/addresses.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"} # these are the Header's
req = requests.get(url, headers=headers)
data = StringIO(req. text)
df2 = pd.read_csv(data)
```

## Index_col parameter
* chaning a column into index

```python
pd.read_csv("daily_and_month_call_report.csv",index_col="Month")
```

## Header Parameter (important)
* what if header is merged in data

```python
pd.read_csv("daily_and_month_call_report.csv",header=1)
```

## use_cols Parameter
* somethimes we don't need extra columns so we can remove them on time of importing

```python
pd.read_csv("daily_and_month_call_report.csv", usecols=['Month',"Year"])
```

## Squeeze parameters

```python
pd.read_csv("daily_and_month_call_report.csv",usecols=['Month'],squeeze=True )
```

## Skipping/nrows Parameter
* Skipping the rows which we don't Need
* we can make a fuction and pass here that remove all the multiple rows of 5 something like that will be good 

```python
pd.read_csv("daily_and_month_call_report.csv", skiprows=[4,1])
```

## nrows

```python
pd.read_csv("daily_and_month_call_report.csv", nrows=100)
```

## Encoding Parameters
* Sometime on kaggle data don't have UTF-8  how to solve that
* First way is to convet that file into UTF-8
* Secondly we can do find encoding type of that file and use encoding="Type"

```python
pd.read_csv("Mumbai Local Train Dataset.csv", encoding='latin-1')
```

## Skip Badlines
* sometimes some rows have extra values it will throw parseerror means having extra valus in rows than expected 

```python
pd.read_csv("Mumbai Local Train Dataset.csv", encoding='latin-1', error_bad_lines=False) # it will skip the row which containes error
```

## Dtype parameter
* Below we are converting int -> float 

```python
pd.read_csv("daily_and_month_call_report.csv",dtype={"Number of Total Health Information": float})
```

## Handling Dates
* Basically sometimes in data dates are given in object(sting) we can change that using parse_dates=['dates']

```python
pd.read_csv("countryLockdowndates.csv",parse_dates=['dates']) # chage to datetime64[ns]
```

```python
def rename(name):
    if name == 'Aug':
        return "Augest"
    else:
        return name
```

## Convertors
* Powerfull sometimes we have 

```python
pd.read_csv("daily_and_month_call_report.csv", converters={'Month':rename})
```

## na_values parameter

```python
pd.read_csv("countryLockdowndates.csv",na_values=['Dec'])
```
